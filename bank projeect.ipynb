### que : - Which customer profiles job have higher subscription rates?

```python
a=df.groupby('job')['y'].value_counts(normalize=True).unstack()*100
a
```

```python
a.plot(kind='bar', stacked=True,figsize=(10,5), color=['red', 'black'], edgecolor='black')
plt.title('Subscription by Job Type')
plt.xlabel('Job')
plt.ylabel('Count')
plt.legend(['Not Subscribed (0)', 'Subscribed (1)'],bbox_to_anchor=(1.0,1.02))
plt.xticks(rotation=45)
plt.show()
```

# Previous Campaign Impact:

### que : - How does the outcome of previous campaigns influence the current one?


```python
i = df.groupby('poutcome')['y'].value_counts(normalize=True).unstack(fill_value=0) * 100
i.columns = ['No (0)', 'Yes (1)']
i = i.sort_values(by='Yes (1)', ascending=False)
```

```python
i.plot(kind='bar', figsize=(10, 6), color=['red', 'black'], edgecolor='black')
plt.title('Subscription Rate by Outcome of Previous Campaigns')
plt.xlabel('Previous Campaign Outcome (poutcome)')
plt.ylabel('Percentage')
plt.legend(title='Subscription',bbox_to_anchor=(1.15,1.02))
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--')
plt.show()
```

# Temporal Analysis

### que - Which months have higher subscription rates?

```python
k = df.groupby('month')['y'].value_counts(normalize=True).unstack(fill_value=0) * 100
month_order = ['jan', 'feb', 'mar', 'apr', 'may', 'jun',
               'jul', 'aug', 'sep', 'oct', 'nov', 'dec']
k = k.reindex(month_order)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(k.index, k[1], marker='o', color='black', linewidth=2)
plt.title('Trend of Subscription Rate Over Months')
plt.xlabel('Month')
plt.ylabel('Subscription Rate (%)')
plt.grid(linestyle='--')
plt.xticks(rotation=45)
plt.show()
```

### que : - Is there any effect of the day of the week on subscription success?

```python
l = df.groupby('day_of_week')['y'].value_counts(normalize=True).unstack(fill_value=0) * 100
weekday_order = ['mon', 'tue', 'wed', 'thu', 'fri']
l = l.reindex(weekday_order)
```

```python
l.plot(kind='bar', figsize=(10, 6), color=['red', 'black'], edgecolor='black')

plt.title('Subscription Rate by Day of the Week')
plt.xlabel('Day of Week')
plt.ylabel('Percentage')
plt.legend(['No (0)', 'Yes (1)'],title='Subscription',bbox_to_anchor=(1.0,1.02))
plt.grid(axis='y', linestyle='--')
plt.show()
```
